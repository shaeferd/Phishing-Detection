```python
def create_mail_df(my_dir):
    titles = []
    contents = []
    labels = []

    for f in os.listdir(os.path.join('phish',my_dir)):
            with open(os.path.join('phish', my_dir, f), 'r') as reader:
                try:
                    c = reader.read()
                except:
                    continue
                contents.append(c)
                #titles.append(f)
                labels.append('phish')

    df = pd.DataFrame({'content': contents, 'label': 'phish'},
                            columns = ['label', 'content'])
    return df
```
